```python
# inscriptions which were produced at the year 165 at the earliest
turning_point = 165
len(EDHCS[EDHCS["not_before"] >= turning_point])
```

```python
%%time
turning_point = 165

cities_occups_data = []
for city in cities_latin["Ancient Toponym"].tolist():
    city_hanson = cities_hanson[cities_hanson["Ancient Toponym"]==city]
    edhcs_city_df = EDHCS[EDHCS["nearest_city"]==city]
    ins_N = len(edhcs_city_df)
    try:
        ins_N_turn = len(edhcs_city_df[edhcs_city_df["not_before"] >= turning_point]) / ins_N
    except:
        ins_N_turn = 0
    occups_lists = [eval(el) for el in edhcs_city_df["occups"].tolist()]
    occups_unique_list = list(set([el for sublist in occups_lists for el in sublist]))
    occups_unique_N = len(occups_unique_list)
    occups_N = edhcs_city_df["occups_N"].sum()
    city_data = {
        "city" : city,
        "pop_est" : city_hanson["pop_est"].tolist()[0], 
        "Country" : city_hanson["Country"].tolist()[0], 
        "geometry" : city_hanson["geometry"].tolist()[0],
        "ins_N" : ins_N,
        "ins_N_turn" : ins_N_turn,
        "unique_occups" : occups_unique_list,
        "occups_N" : occups_N,
        "unique_occups_N" : occups_unique_N
    }
    cities_occups_data.append(city_data)
cities_occups_df = pd.DataFrame(cities_occups_data)
cities_occups_gdf = gpd.GeoDataFrame(cities_occups_df)
```

```python
cities_occups_gdf["occups_pairs"] = cities_occups_gdf["unique_occups"].apply(lambda items: [(items[i],items[j]) for i in range(len(items)) for j in range(i+1, len(items))])
```

```python
cities_occups_filt_gdf = cities_occups_gdf[cities_occups_gdf["unique_occups_N"] >= 2]
occups_pairs_list = cities_occups_filt_gdf["occups_pairs"].tolist()
```

For each city, we extract all unique occupation names appearing on inscriptions from that city. Subsequently, we generate all possible combinations or pairs of these occupations in the given city and evaluate how often each individual pair of occupations appears  in all the other cities in the dataset.

If a pair of occupations (*i*,*j*) appears only in the given city, it has a score equal to 1, which is a maximum value.
 If the same pair appears in more than one city, the score is decreased proportionally as:

$\frac{1}{N of cities with (i,j)}$

The *tightness score* of given city is then calculated as an avarege of all scores of all occupations' pairs.

For instance, in Londinium, there are only 2 unique occupations: *scriba* (a scribe/writer) and *cuparius* (a cooper). Since these two occupations do not appear together in any other city, the tightness score is equal to 1. On the other hand, in Nemausus, there are only 2 unique occupations as well: *utricularius* (a bagpiper) and *faber* (a worker). However, this is a more common combination, appearing altogether in 4 different cities. Accordingly, the *tightness score* equals to 0.25.

```python
# test with Rome
for pair in occups_pairs_list[0]:
    if len([pairs_list for pairs_list in occups_pairs_list if pair in pairs_list]) > 1:
        print(pair, len([pairs_list for pairs_list in occups_pairs_list if pair in pairs_list]))
        #print(occups_pair_freq)
```

```python
all_pairs = []
cities_scores = []
for pairs in occups_pairs_list:
    city_pair_scores = []
    for pair in pairs:
        all_pairs.append(pair)
        pair_score = 1 / len([pairs_list for pairs_list in occups_pairs_list if pair in pairs_list])
        city_pair_scores.append(pair_score)
    city_score = np.mean(city_pair_scores)
    cities_scores.append(city_score)
```

```python
cities_occups_filt_gdf["tightness_score"] = cities_scores
```
